# Imports

```python
from transformers import pipeline
import pandas as pd
```

```python
tqa = pipeline(task="table-question-answering", 
               model="google/tapas-base-finetuned-wtq")
```

# Reading Data

```python
table = pd.read_csv(r"C:\Users\ShubhamGupta\Downloads\Data.csv",nrows=30)
table = table.astype(str)
```

```python
table
```

```python
def process_string(input_string):
    words = input_string.split(' ')
    first_word = words[0]

    if first_word == 'COUNT':
        count = input_string.count(',')
        return count
    elif first_word == 'SUM':
        numbers = [float(s) for s in input_string.split() if s.replace('.', '', 1).isdigit()]
        return sum(numbers)
    elif first_word == 'AVERAGE':
        numbers = [float(s) for s in input_string.split() if s.replace('.', '', 1).isdigit()]
        return sum(numbers) / len(numbers) if numbers else 0
    else:
        return input_string
```

```python
query = "Whose age is the highest?"
str = tqa(table=table, query=query)["answer"]
# print(str)
print(process_string(str))
```
